```python
# ðŸ“Œ Step 0 â€” Environment setup and imports  
import sys
import os
import clifpy
from pathlib import Path

from clifpy import Labs
```

```python
# ðŸ“Œ Step 1 â€” Load the Labs table  
# The `Labs.from_file` helper reads the Parquet files, attaches the DataFrame to  
# `labs.df`, and sets up logging & schema information automatically.

# Find the demo lab table shipped with the package
example_labs_dir = Path(clifpy.__file__).parent / "data" / "clif_demo"  

labs = Labs.from_file(
    data_directory=str(example_labs_dir),
    filetype="parquet",
    timezone="UTC"
)
```

```python
# ðŸ“Œ Step 2 â€” Reference Units Demo  
# Check and display lab reference units loaded from the schema (via Labs class property).
# This demonstrates how to access the mapping from lab test names to their reference units.
lab_reference_units = labs.lab_reference_units  # This property reads from the schema in Labs
print("Sample of available lab reference units from package schema:")
for lab, units in list(lab_reference_units.items())[:20]:  # show first 5 entries as a sample
    print(f"{lab}: {units}")
```
