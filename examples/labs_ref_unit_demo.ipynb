```python
# 📌 Step 1 — Load the Labs table  
# The `Labs.from_file` helper reads the Parquet files, attaches the DataFrame to  
# `labs.df`, and sets up logging & schema information automatically.

# Find the demo lab table shipped with the package
example_labs_dir = Path(clifpy.__file__).parent / "data" / "clif_demo"  

labs = Labs.from_file(
    data_directory=str(example_labs_dir),
    filetype="parquet",
    timezone="UTC"
)
```

```python
# 📌 Step 2 — Reference Units Demo  
# Check and display lab reference units loaded from the schema (via Labs class property).
lab_reference_units_df = labs.get_lab_reference_units(save=True)
lab_reference_units_df
```

```python
# Standardize
labs.standardize_reference_units(save=True, lowercase=True)

# Check units after
print("\nAfter:")
print(labs.get_lab_reference_units())
```

```python
# Process
labs.process_reference_units(lowercase=True)

# Check units after
print("\nAfter:")
print(labs.get_lab_reference_units())
```
